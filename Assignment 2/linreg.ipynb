```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
abalone_df = pd.read_csv('abalone.txt', delimiter = ',')
abalone_df.head()
```

```python
abalone_feature = ['Length', 'Diameter','Height', 'Whole weight', 
                   'Shucked weight', 'Viscera weight', 'Shell weight'
                  ]

plt.figure(figsize = (15,10))
for i in enumerate(abalone_feature):
    # plt.subplot 3 rows, 3 columns, plot i+1 since plot must start at 1
    plt.subplot(3, 3, i[0]+1)
    sns.scatterplot(data=abalone_df, x=i[1], y='Rings')
```
